```python
import json

with open("aquinas_new.json", "r") as handle:
    data = json.load(handle)
data[0]
```

```python
import nltk
#nltk.download('wordnet')

lemma = nltk.wordnet.WordNetLemmatizer()


from collections import defaultdict, Counter
import spacy
nlp = spacy.load('en')
import tqdm

kill = set(['it', 'i', 'it', 'that', 'hath', 'thou', 'hast', 'that', 'who', 'it', 'which',
        'reason', 'it','we', 'which', 'which', 'it', 'which', 'this', 'others',
       'things', 'things', 'things', 'they', 'what', 'he', 'most', 'words', 'word', 'something', 'them', 'thing',
            'inasmuch', 'some', 'these', 'whence', 'whatever', '1', 'whoever', 'reply', 'both'
       ])

article_topics = {}
n = len(data)
for i, article in enumerate(data):
    if i % n == 0:
        print("%{} complete".format((i/n)*100))
    subjects = []
    for j, sentence in enumerate(nltk.sent_tokenize("\n".join([k for k in article["articleBody"]]))):
        doc = nlp(sentence)
        for word in doc:
            if word.dep_ == "nsubj":
                if not word.text.lower() in kill:
                    lemmatized_word = lemma.lemmatize(word.text.lower())

                    subjects.append(lemmatized_word)
    
    #dict(Counter(subjects).most_common())
    article_topics[i] = dict(Counter(subjects).most_common())
    #print("article{}".format(i), subjects)
        

    
```
